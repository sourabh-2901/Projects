```python
avg_school = df.groupby("School")["Marks"].mean().sort_values()

avg_school.plot(kind='barh', color='skyblue', figsize=(8, 4))
plt.title("Average Marks by School")
plt.xlabel("Average Marks")
plt.ylabel("School")
plt.tight_layout()
plt.show()
```

```python
top_students = df.groupby(["Name"])["Marks"].sum().sort_values(ascending=False).head(5)

top_students.plot(kind='barh', color='orange', figsize=(8, 4))
plt.title("Top 5 Students (Total Marks)")
plt.xlabel("Total Marks")
plt.gca().invert_yaxis()  # Highest at top
plt.tight_layout()
plt.show()
```

```python
heat_data = df.groupby(["School", "Subject"])["Marks"].mean().unstack()

plt.figure(figsize=(10, 5))
sns.heatmap(heat_data, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title("Avg Marks per Subject per School")
plt.tight_layout()
plt.show()
```

```python
df["Pass"] = df["Marks"] >= 33
pass_count = df["Pass"].value_counts()

labels = ["Passed", "Failed"]
colors = ["#4CAF50", "#F44336"]
plt.pie(pass_count, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title("Overall Pass/Fail Distribution")
plt.axis('equal')  # Circle
plt.show()
```

```python
subject_avg = df.groupby("Subject")["Marks"].mean()
colors = ["red" if avg < 40 else "green" for avg in subject_avg]

subject_avg.plot(kind='bar', color=colors)
plt.title("Subject-Wise Average Marks (Highlighting Weak Areas)")
plt.ylabel("Average Marks")
plt.axhline(40, color='gray', linestyle='--', label="Pass Threshold")
plt.legend()
plt.tight_layout()
plt.show()
```
